```python
import re
# regex , which can be used to work with Regular Expressions
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import string
import nltk
import warnings 
warnings.filterwarnings("ignore", category=DeprecationWarning)

%matplotlib inline
```

```python
#Combine the train.csv and test.csv files
combine = df.append(test,ignore_index=True,sort=True)
```

```python
#Removing Twitter Handles(@User)
def remove_pattern(text,pattern):
    
    # re.findall() finds the pattern i.e @user and puts it in a list for further task
    r = re.findall(pattern,text)
    
    # re.sub() removes @user from the sentences in the dataset
    for i in r:
        text = re.sub(i,"",text)
    
    return text
```

```python
combine
```

```python
"""Define a vectorized function which takes a nested sequence of objects or numpy arrays as inputs and returns
      a single numpy array 
    or a tuple of numpy arrays.
The vectorized function evaluates pyfunc over successive tuples of the input arrays like the python map function,
           except it uses the broadcasting rules of numpy."""

combine['Tidy_Tweets'] = np.vectorize(remove_pattern)(combine['tweet'], "@[\w]*")

combine.head(10)
```

```python
#Removing Punctuation, Numbers, and Special Characters

"""string method replace() returns a copy of the string in which the occurrences of old have been replaced with new,
     optionally restricting the number of replacements to max."""

combine['Tidy_Tweets'] = combine['Tidy_Tweets'].str.replace("[^a-zA-Z#]", " ")

combine.head(10)
```

```python
#Removing Short Words

combine['Tidy_Tweets'] = combine['Tidy_Tweets'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))

combine.head(10)
```

```python
"""Now we will tokenize all the cleaned tweets in our dataset. Tokens are individual terms or words,
and tokenization is the process of splitting a string of text into tokens"""

tokenized_tweet = combine['Tidy_Tweets'].apply(lambda x: x.split())

tokenized_tweet.head()
```

```python
#Stemming is a rule-based process of stripping the suffixes (“ing”, “ly”, “es”, “s” etc) from a word
from nltk import PorterStemmer

ps = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda x: [ps.stem(i) for i in x])

tokenized_tweet.head()
```

```python
#Now let’s stitch these tokens back together
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])

combine['Tidy_Tweets'] = tokenized_tweet
combine.head()
```

```python
#Function to extract hashtags from tweets
def Hashtags_Extract(x):
    hashtags=[]
    
    # Loop over the words in the tweet
    for i in x:
        ht = re.findall(r'#(\w+)',i)
        hashtags.append(ht)
    
    return hashtags
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bow_vectorizer = CountVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words='english')

# bag-of-words feature matrix
#Fit to data, then transform it.
bow = bow_vectorizer.fit_transform(combine['Tidy_Tweets'])
#todense returns a matrix
"""todense. A NumPy matrix object with the same shape and containing the same data represented by the sparse matrix
with the requested memory order"""
df_bow = pd.DataFrame(bow.todense())

df_bow
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#TF-IDF weight is composed by two terms
tfidf=TfidfVectorizer(max_df=0.90, min_df=2,max_features=1000,stop_words='english')

tfidf_matrix=tfidf.fit_transform(combine['Tidy_Tweets'])

df_tfidf = pd.DataFrame(tfidf_matrix.todense())

df_tfidf.head(10)
```

```python
train_bow = bow[:31962]

train_bow.todense()
```

```python
#Using features from TF-IDF for training set
train_tfidf_matrix = tfidf_matrix[:31962]

train_tfidf_matrix.todense()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Bag-of-Words Features
x_train_bow, x_valid_bow, y_train_bow, y_valid_bow = train_test_split(train_bow,df['label'],test_size=0.3,random_state=2)
```

```python
#TF-IDF features
x_train_tfidf, x_valid_tfidf, y_train_tfidf, y_valid_tfidf = train_test_split(train_tfidf_matrix,df['label'],test_size=0.3,random_state=17)
```

```python
from sklearn.metrics import f1_score
```

```python
#The first model we are going to use is Logistic Regression.
from sklearn.linear_model import LogisticRegression
Log_Reg = LogisticRegression(random_state=0,solver='lbfgs')
```

```python
#Bag-of-Words Features
#Fitting the Logistic Regression Model.
Log_Reg.fit(x_train_bow,y_train_bow)
```

```python
#Predicting the probabilities.
prediction_bow = Log_Reg.predict_proba(x_valid_bow)

prediction_bow
```

```python
#Calculating the F1 score
# if prediction is greater than or equal to 0.3 than 1 else 0
# Where 0 is for positive sentiment tweets and 1 for negative sentiment tweets
prediction_int = prediction_bow[:,1]>=0.3

# converting the results to integer type
prediction_int = prediction_int.astype(np.int)
prediction_int

# calculating f1 score
log_bow = f1_score(y_valid_bow, prediction_int)

log_bow
```
